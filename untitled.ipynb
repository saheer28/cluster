```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/dataset/Mall_Customers.csv')
df
```

```python
X=df.iloc[:,2:]
X
```

```python
from sklearn.preprocessing import normalize
data=normalize(X)
data
```

```python
newdata=pd.DataFrame(data,columns=X.columns)
newdata
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.title('DENDROGRAM')
dend=sch.dendrogram(sch.linkage(newdata,method='complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
cluster.fit_predict(newdata)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'],c=cluster.labels_)
```
